```python
import pandas as pd 
import numpy as np 
import folium
import csv
import math 

#list of country name, lat / long , attack type, target, nkill, nkill ter, n wound, nwound us, nwound te
data = pd.read_csv("GTD2013_2016.csv", encoding = "utf-8")
country_freq = {}
country_lat = {}
country_long = {}
country_lat_list = []
country_long_list = []
country_year = []
country_month = []




```

```python
#find average long / lat for each country
for i in range(0, len(data)):
    print(i)
    if data.iloc[i] ["country_txt"] in country_lat:
        if not math.isnan(data.iloc [i]["latitude"]) and not math.isnan(data.iloc [i]["longitude"]):
            country_lat[data.iloc[i] ["country_txt"]] = country_lat[data.iloc[i] ["country_txt"]] + data.iloc [i]["latitude"]
            country_long[data.iloc[i] ["country_txt"]] = country_long[data.iloc[i] ["country_txt"]] + data.iloc [i]["longitude"]
            country_freq[data.iloc[i] ["country_txt"]] = country_freq[data.iloc[i] ["country_txt"]] + 1
            country_year.append(data.iloc[i]["iyear"])
            country_lat_list.append(data.iloc[i]["latitude"])
            country_long_list.append(data.iloc[i]["longitude"])
            country_month.append(data.iloc[i]["imonth"])
    else: 
        if not math.isnan(data.iloc [i]["latitude"]) and not math.isnan(data.iloc [i]["longitude"]):
            country_lat[data.iloc[i] ["country_txt"]] = data.iloc [i]["latitude"]
            country_long[data.iloc[i] ["country_txt"]] = data.iloc [i]["longitude"] 
            country_freq[data.iloc[i] ["country_txt"]] = 1
            country_year.append(data.iloc[i]["iyear"])
            country_lat_list.append(data.iloc[i]["latitude"])
            country_long_list.append(data.iloc[i]["longitude"])
            country_month.append(data.iloc[i]["imonth"])
```

```python
for i in country_lat:
    country_lat[i] =  country_lat[i] / country_freq[i]
    country_long[i] = country_long[i] / country_freq[i]
print (country_lat)
```

```python
import pandas as pd 
import numpy as np 
import folium
import csv
import math 
import matplotlib.pyplot as plt
import re

date_hours = r"^\d\d?\/(\d\d?)\/18 (\d\d?):.*$"
western_data = pd.read_csv("western.csv", usecols= ["created_at", "retweet_count"], encoding = "latin1")
eastern_data = pd.read_csv("eastern.csv", usecols= ["created_at", "retweet_count"], encoding = "latin1")
data = {}

def date(day, is_east):
    data = {}
    if is_east:
        data_frame = eastern_data
    else:
        data_frame = western_data
    
    for i in range(0, len(data_frame)):
        matchObj = re.match(date_hours, str(data_frame.iloc[i]["created_at"]))
        if matchObj:
            date = int(matchObj.group(1))
            hour = int(matchObj.group(2)) 
            recount = data_frame.iloc[i]["retweet_count"]
            if (date == day) and not math.isnan(float(recount)):
                if hour in data:
                    data[hour] += int(recount)
                else:
                    data[hour] = int(recount)    
        else:
            print ("No match!!")
    
    return data

data_2_W = date(2, False)
data_3_W = date(3, False )
data_4_W = date(4, False)
data_5_W = date(5, False)
data_6_W = date(6, False)
data_7_W = date(7, False)
data_8_W = date(8, False)
data_9_W = date(9, False)
data_10_W = date(10, False)

data_2_E = date(2, True)
data_3_E = date(3, True)
data_4_E = date(4, True)
data_5_E = date(5, True)
data_6_E = date(6, True)
data_7_E = date(7, True)
data_8_E = date(8, True)
data_9_E = date(9, True)
data_10_E = date(10, True)

print(data_10_W)
print(data_2_E)
```

```python
def plot_bar(data, title, count):
    # this is for plotting purpose
    plt.figure(count)
    y_pos = np.arange(len(data))
    plt.bar(y_pos, data.values())
    plt.xlabel('Hour', fontsize=10)
    plt.ylabel("Number of Retweets", fontsize = 10)
    plt.xticks(y_pos, data.keys(), fontsize=10, rotation=30)
    plt.title(title)
    plt.show()

plot_bar(data_5_W, "Retweets May 5th Western World", 1)
plot_bar(data_5_E, "Retweets May 5th Eastern World", 2)

plot_bar(data_6_W, "Retweets May 6th Western World", 3)
plot_bar(data_6_E, "Retweets May 6th Eastern World", 4)

plot_bar(data_7_W, "Retweets May 7th Western World", 5)
plot_bar(data_7_E, "Retweets May 7th Eastern World", 6)

plot_bar(data_8_W, "Retweets May 8th Western World", 7)
plot_bar(data_8_E, "Retweets May 8th Eastern World", 8)

plot_bar(data_9_W, "Retweets May 9th Western World", 9)
plot_bar(data_9_E, "Retweets May 9th Eastern World", 10)

plot_bar(data_10_W, "Retweets May 10th Western World", 11)
plot_bar(data_10_E, "Retweets May 10th Eastern World", 12)
```
